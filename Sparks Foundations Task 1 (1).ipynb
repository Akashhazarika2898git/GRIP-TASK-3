```python
samplestore1_df = samplestore_df[['City','Quantity']]  ## here in code we extracted the desired columns for our analysis process 
```

```python
samplestore1_df
```

```python
samplestore1_df = pd.DataFrame(samplestore_df.groupby(['City'])['Quantity'].sum()) ## here we use groupby DataFrame for  mapping of columns 
```

```python
samplestore1_df.reset_index(inplace = True) ## this is used to reset the present index 
```

```python
samplestore1_df= samplestore1_df.sort_values(['Quantity'], ascending = False).head(10) ## sortvalues() is used to sort the values either in ascending or descending 
```

# Data Visualization

```python
import matplotlib.pyplot as plt ## importing matplot.pyplot library 
```

```python
plt.figure(figsize = (15,8))  ## this is ud=sed to adjust the our plot size 
plt.plot(samplestore1_df['City'], samplestore1_df['Quantity']) ## plot is used to visualize our pattern between the columns in our data frame 
```

```python
plt.figure(figsize = (15,8))
plt.bar(samplestore1_df['City'], samplestore1_df['Quantity'],color = 'g') ## .bar is used to get plot of type bar
```

```python
samplestore_df['Category'].value_counts()
```

```python
plt.pie(samplestore_df['Category'].value_counts(), labels = samplestore_df['Category'].value_counts().index, autopct = '%1.1f%%', radius= 2)
```

# Importing Seaborn library 

```python
import seaborn as sns ## importing seaborn library for our data analysis 
```

```python
samplestore2_df  = pd.DataFrame(samplestore_df.groupby(['Category'])['Sub-Category'].value_counts())
```

```python
plt.figure(figsize = (15,8))
sns.heatmap(samplestore2_df) ## This is used to get heatmap that is 2-dimensional figure to visualize  
```

```python
samplestore3_df = pd.DataFrame(samplestore_df.groupby(['Sub-Category'])['Sales'].sum())
```

```python
samplestore3_df
```

```python
plt.figure(figsize = (20,8))
plt.bar(samplestore3_df.index, samplestore3_df['Sales'], color = 'r')
```

```python
samplestore4_df = pd.DataFrame(samplestore_df.groupby(['Sub-Category'])['Profit'].sum())
```

```python
samplestore4_df
```

```python
plt.figure(figsize = (20,8))
plt.bar(samplestore4_df.index, samplestore4_df['Profit'], color = 'orange')
```
